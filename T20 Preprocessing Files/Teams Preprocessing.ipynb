```python
import json

# Specify the complete file path
file_path = r'D:\Cricket Analytics Project\t20_json_files\t20_wc_match_results.json'

# Open and read the JSON file
with open(file_path, 'r') as json_file:
    data = json.load(json_file)

# Print or work with the JSON data
##data

#converting to DataFrame
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
```

```python
df_match.head()
```

## Creating match ids to connect with the Batting Summary

```python
match_ids_dict={}
for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs '+ row['team2']
    key2 = row['team2'] + ' Vs '+ row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
file_path = r'D:\Cricket Analytics Project\t20_json_files\t20_wc_batting_summary.json'

# Open and read the JSON file
with open(file_path, 'r') as json_file:
    data = json.load(json_file)

# Print or work with the JSON data
data

#joining other batting summaries togetherwe are using extend
allrecords = []
for rec in data:
    allrecords.extend(rec['battingSummary'])
df_batting = pd.DataFrame(allrecords)
df_batting.head()
```

```python
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x : 'out' if len(x) > 0 else 'not out')
```

```python
df_batting
```

```python
df_batting.drop(['dismissal'],axis=1,inplace=True)
```

```python
df_batting
```

```python
df_batting['match_ids'] = df_batting['match'].map(match_ids_dict)
```
